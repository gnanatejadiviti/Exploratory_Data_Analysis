```python
lower       upper        frequency
2.136        7.98          2992
```

```python
fre,inter,n=plt.hist(visa_df['prevailing_wage'],
        bins=40)
```

```python
frequency = []
index= []
for i in range(len(n)):
    con1 = visa_df['prevailing_wage']>=inter[i]
    con2 = visa_df['prevailing_wage']<=inter[1:][i]
    c = con1 & con2
    frequency.append(len(visa_df[c]))
    index.append(i+1)
```

```python
df = pd.DataFrame(zip(inter,inter[1:],frequency),columns=['Lower','Upper','Frequency'],index=index)
df
```

## Boxplot
- **Boxplot is used to identify outliers**
- **In box plot we have**
    - **Q1**: 25p value
    - **Q2**: 50p value
    - **Q3**: 75p value
    - **IQR**: Q3-Q1
    - Mild outliers Q1-1.5xIQR and Q3+1.5xIQR
    - Huge outliers Q1-3xIQR and Q3+3xIQR

```python
plt.boxplot(visa_df['prevailing_wage'],vert=False)
plt.show()

# Black dots are outliers
# Orange line is median
# vert: False (Horizontal), vert: True (Vertical,default)
```
