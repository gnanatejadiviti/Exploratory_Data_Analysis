#### Create dataframes unsing List

```python
name = ['Ramesh','Suresh','Sathish']
age = [30,35,40]
name,age
```

**Step-1**\
$Create$ $dataframe$

```python
data = (zip(name,age))
data
```

**Step-2**\
$Providing$ $the$ $columns$ $and$ $index$

```python
df = pd.DataFrame(data,columns=['Name','Age'],index= ['A','B','C'])
df
```

**Step-3**\
$Add$ $new$ $column$
- If you want to add a new column
- df['new_column_name']
- You need to have a list which is having some elements
- That elements need to equal to number of rows
  

```python
city_names = ['Hyderabad','Mumbai','Chennai']
df['City'] = city_names
df
```

**Step-4**\
$Updating$ $the$ $existing$ $column$

```python
df['Name']=['Teja','Gnana','Ashish']
df
```

**Step-5**\
$Drop$ $the$ $column$
- In order to drop the column
- We need to use drop method
- It takes 3 parameters
     - Drop column or row
     - mention the coulmn name
     - axis
          - axis = 1 reference as column
          - axis = 0 reference as row
     - you want to create a new dataframe or
     - you want to overwrite the existing dataframe
          - inplace = True

```python
df.drop('C',
       axis = 0,
       inplace=True)
df
```

```python
df.drop('Name',
       axis = 1,
       inplace= True)
df
```

```python
name = ['Ramesh','Suresh','Sathish']
age = [30,35,40]

data = (zip(name,age))

df = pd.DataFrame(data,columns=['Name','Age'],index= ['A','B','C'])

city_names = ['Hyderabad','Mumbai','Chennai']
df['City'] = city_names

df.drop('C',
       axis = 0,
       inplace = True)
df
```

**Step-6**\
$Saving$ $the$ $data$ $frame$

```python
df.to_csv("Dataframe1.csv")
```
