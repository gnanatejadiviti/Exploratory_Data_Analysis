$take-lock-iloc$
#### If you want specific rows or coulmns

```python
df.take([2,4,6])
# By default axis is 0 i.e, Rows
#            axis is 1 i.e, Columns
```

```python
df.take([0,1,6,8,2],axis=1)
```

```python
df1 = df.take([100,200,300])
df1.take([4,8,11],axis=1)
#################################OR#####################################
df.take([100,200,300]).take([4,8,11],axis=1)
```

**Note**
- Take does not take rows and columns at a time

$iloc$

```python
# df.iloc[<rows>,<colums>]
# df.iloc[<start:end>,<start:end>]    (For specific data)
```

```python
df.iloc[5:10] # All the columns
```

```python
df.iloc[5:10,2:12]  # Specific rows and specific columns
```

```python
df.iloc[:,2:5] # All the rows and 2 to 4 columns
```

```python
df.iloc[[100,200,300],[4,8,11]]
```

```python
df.iloc[[100,200,300],8]
# No brackets: Series
# Bracket : Dataframe
```

```python
df.iloc[[100,200,300],[8]]
```

**Note**
- In iloc we cant provide the name of the column
